```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

## Load background model

```python
def deserializeMatbin(filename):
    
    f = open(filename, "rb")
    cols = np.fromfile(f, dtype=np.uint32, count=1)[0]
    rows = np.fromfile(f, dtype=np.uint32, count=1)[0]
    elemSizeInBytes = np.fromfile(f, dtype=np.uint32, count=1)[0]
    elemType = np.fromfile(f, dtype=np.uint32, count=1)[0]
    
    print("cols:", cols, "rows:", rows, "elemSizeInBytes:",
          elemSizeInBytes, "elemType:", elemType)
    
    output = np.fromfile(filename, dtype='i2', count=cols * rows).reshape(rows, cols, 1)

    return output

bg_model16b = deserializeMatbin("/home2/aifi_code/depth/bg_sub/build/depth_model_aligned.matbin")
print("depth16b:", bg_model16b.shape)
print("depth16b:", bg_model16b.dtype)
print("Median val:", np.median(bg_model16b))
print("Min val:", np.amin(bg_model16b), "Max val:", np.amax(bg_model16b))
```

```python
# Convert to 8b for visualization
bg_model8b = np.divide(bg_model16b, 16)
print("Median val:", np.median(bg_model8b))
print("Min val:", np.amin(bg_model8b), "Max val:", np.amax(bg_model8b))

min_val = np.amin(bg_model8b)
max_val = np.amax(bg_model8b)
print("min val:", min_val)
print("max val:", max_val)

plt.imshow(bg_model8b[:, :, 0], cmap='gray')
plt.title('Depth8b')
plt.show()
```

## Load an image

```python
img16b = deserializeMatbin("/fast_data/top_view_mod/vid2/199d16.matbin")
print("img16b:", img16b.shape)
img8b = np.divide(img16b, 16)
min_val, max_val = np.amin(img8b), np.amax(img8b)
print("min val:", min_val)
print("max val:", max_val)

plt.imshow(img8b[:, :, 0], cmap='gray')
plt.title('Depth8b')
plt.show()
```

## Segment the background

```python
# Subtract the background model
def remove_background(img, bg_model, is8bit = False):
    if is8bit:
        thresh1 = 25
        thresh2 = 2000.0
        mask_type = np.uint8
    else:
        thresh1 = 400
        thresh2 = 2000.0
        mask_type = np.int16
    
    kernel = np.ones((5,5),np.uint8)
    forg = bg_model - img;
    ret, forgbin = cv2.threshold(forg, thresh1, 255, cv2.THRESH_BINARY)
    forgbin = forgbin.astype(np.uint8)

    img, contours, h = cv2.findContours(forgbin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    del_idxs = []
    for i, c in enumerate(contours):
        area = cv2.contourArea(c)

        if area < thresh2:
            del_idxs.append(i)

    contours = [i for j, i in enumerate(contours) if j not in del_idxs]

    mask = np.zeros((480, 640), mask_type)
    cv2.drawContours(mask, contours, -1, 1, -1);
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # Apply mask
    forg_final = img * mask
    forg_final = forg_final.squeeze()
    
    return forg_final

forg_final = remove_background(img8b, bg_model8b, is8bit = True)

# Depth to 8 bit for visualization
#forg8u = np.divide(forg_final, 16)
    


plt.imshow(forg_final[:, :], cmap='gray')
plt.title('Depth8b')
plt.show()
```
